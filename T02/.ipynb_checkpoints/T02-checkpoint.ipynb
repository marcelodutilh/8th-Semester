```python
from collections import defaultdict
conexiones_infeccion = open("conexiones_infecciÃ³n.csv","r")
conexiones_viaje = open("conexiones_viaje.csv","r")
paises = open("paises.csv","r")


def mapa_viaje(archivo_viaje):
    mapa_viaje = list()
    for linea in archivo_viaje:
        if "Pais 1" in linea:
            pass
        else:
            line = linea.replace("\n","")
            line = line.split(";")
            mapa_viaje.append(line)
    
    lista_paises = set()
    for conexion in mapa_viaje:
        lista_paises.add(conexion[0])
        lista_paises.add(conexion[1])
    
    mapa = defaultdict(dict)
    for pais in lista_paises:
        conexiones = defaultdict(dict)
        for conexion in mapa_viaje:
            if conexion[0] == pais:
                z = {conexion[1] : 1}
                conexiones[pais].update(z)
        mapa.update(conexiones)
    return mapa

mapa_viaje = mapa_viaje(conexiones_viaje)  
conexiones_viaje.close()
conexiones_viaje = open("conexiones_viaje.csv","r")

class Tierra:
    
    def __init__(self):
        self.paises = {}
    
    def imprimir_mundo(self):
        print("______LISTA DE PAISES______")
        for pais in self.paises:
            print(self.paises[pais].nombre)
        print("______.______")
        
    def iterar_sobre_Mundo(self,T):
        for i in range(T):
            pais_porcentaje = dict()
            for pais in self.paises.values():
                if pais.cantidad_de_infectados_en_T():
                    numero, nombre_pais, new_porcentaje = pais.cantidad_de_infectados_en_T()
                    pais_porcentaje[nombre_pais] = new_porcentaje

            for pais in self.paises.values():
                if pais.nombre in pais_porcentaje.keys():
                    pais.porcentaje_infectados = pais_porcentaje[pais.nombre]
        
    

    @property
    def paso_seguro(self):
        paso_seguro = {}
        no_seguros = []
        for pais in self.paises:
            if self.paises[pais].porcentaje_infectados > 0.2:
                no_seguros.append(self.paises[pais])
        for pais in self.paises:
            if pais in no_seguros:
                pass
            else:
                conexiones_seguras = {}
                print(self.paises[pais].conexiones_viaje)
                for conexion in self.paises[pais].conexiones_viaje:
                    if conexion in no_seguros:
                        pass
                    else:
                        conexiones_seguras.update({conexion:1})
                paso_seguro[pais.nombre] = conexiones_seguras
        print(paso_seguro)
                        
            
        
class Pais:
    
    def __init__(self, nombre, habitantes, tolerancia, subterraneo, porcentaje_infectados):
        self.nombre = nombre
        self.habitantes = int(habitantes)
        self.tolerancia = float(tolerancia)
        self.subterraneo = subterraneo
        self.porcentaje_infectados = float(porcentaje_infectados)/100
        self.conexiones_infeccion = defaultdict(list)
        self.conexiones_viaje = defaultdict(list)
        self.cupula = False
    
    @property
    def tasa_contagio(self):
        x = self.porcentaje_infectados*self.habitantes
        y = (1-(self.porcentaje_infectados))*self.habitantes
        z = self.habitantes
        tasa_contagio = (x-y/100)/z
        
        return tasa_contagio
  

    
    def cantidad_de_infectados_en_T(self):
        
        if not self.cupula and len(self.conexiones_infeccion[self.nombre]) >=1: 
            infectado = self.habitantes * self.porcentaje_infectados
            sanos = self.habitantes * (1-self.porcentaje_infectados)
            tasa_contagio = self.tasa_contagio
            sumatoria_contagios = 0
            Q_vecinos = 0
            promedio_infectados = []
            #cupula
            for vecino in self.conexiones_infeccion[self.nombre]:
                habitantes = vecino.habitantes
                infected = vecino.habitantes * vecino.porcentaje_infectados
                porcentaje = infected/habitantes
                promedio_infectados.append(porcentaje)
            promedio_final = sum(promedio_infectados)/len(promedio_infectados)
            if promedio_final >= 0.7:
                self.cupula = True
                infectados_en_T = infectado + (habitantes-infectado)*self.tasa_contagio
                porcentaje_contagiados_pais = infectados_en_T/self.habitantes

                return infectados_en_T, self.nombre, porcentaje_contagiados_pais
            else:
                #campo fuerza
                for vecino in self.conexiones_infeccion[self.nombre]:
                    condicion_actual = diccionario_conexiones[self.nombre][vecino.nombre]
                    if not condicion_actual:
                        comprobar_nuevo_campo = vecino.tasa_contagio >= self.tolerancia
                        if comprobar_nuevo_campo:
                            diccionario_conexiones[self.nombre][vecino.nombre] = True
                        else:
                            sumatoria_contagios += vecino.tasa_contagio
                            Q_vecinos +=1          
                        
                #print("pais:{}\n infectado:{}\n sanos:{}\n tasa contagio:{}\n sumatoria_cont:{}\n vecinos:{}.".format(self.nombre, infectado,sanos,tasa_contagio,sumatoria_contagios,Q_vecinos))
                if Q_vecinos >= 1:
                    infectados_en_T = infectado + sanos*tasa_contagio + sanos*(1-tasa_contagio)*sumatoria_contagios/(10*Q_vecinos)
                else:
                    infectados_en_T = infectado + sanos*tasa_contagio
                if infectados_en_T >= 0:
                    porcentaje_contagiados_pais = infectados_en_T/self.habitantes
                    return infectados_en_T, self.nombre, porcentaje_contagiados_pais
                else:
                    return 0, self.nombre, 0
        
        elif self.cupula:
            infectado = self.habitantes * self.porcentaje_infectados
            sanos = self.habitantes * (1-self.porcentaje_infectados)
            tasa_contagio = self.tasa_contagio
            infectados_en_T = infectado + (habitantes-infectados)*self.tasa_contagio
            porcentaje_contagiados_pais = infectados_en_T/self.habitantes 
        else:
            return False
        
    def __repr__(self):
        #return "{}, {}, {}, {}, {}".format(self.nombre, self.habitantes, self.tolerancia, self.subterraneo, self.porcentaje_infectados)
        return self.nombre       
def crear_tierra(paises):
    Mundo = Tierra()
    arch_paises = paises
    lista_paises = list()
    for pais in arch_paises:
        new_country = pais.split(";")
        a,b,c,d,e = new_country
        if d == "TRUE":
            j =  True
        else:
            j = False
        f = e.replace("\n","")
        new_country = [a,b,c,j,f]
        lista_paises.append(new_country)
    lista_paises = lista_paises[1:]
    for pais in lista_paises:
        nombre, habitantes, tolerancia, subterraneo, porcentaje_infectados = pais
        country = Pais(nombre, habitantes, tolerancia, subterraneo, porcentaje_infectados)
        Mundo.paises[country.nombre] = country
    return Mundo

def agregar_conexiones_infecciosas(conexiones_infeccion, Mundo):
    lista_infecciones = list()
    arch_infeccion = conexiones_infeccion
    for conexion in arch_infeccion:
        if "Pais 1" in conexion:
            continue
        pais1, pais2 = conexion.split(";")
        pais_1 = pais1.replace("\n","")
        pais_2 = pais2.replace("\n","")
        Mundo.paises[pais_1].conexiones_infeccion[pais_1].append(Mundo.paises[pais_2])

def agregar_conexiones_mapa_viaje(conexiones_viaje, Mundo):
    lista_viaje = list()
    arch_viaje = conexiones_viaje
    for conexion in arch_viaje:
        if "Pais 1" in conexion:
            continue
        pais1, pais2 = conexion.split(";")
        pais_1 = pais1.replace("\n","")
        pais_2 = pais2.replace("\n","")
        Mundo.paises[pais_1].conexiones_viaje[pais_1].append(Mundo.paises[pais_2])
        

#Creo la tierra
World = crear_tierra(paises)

#Agrego las conecciones infecciosas
agregar_conexiones_infecciosas(conexiones_infeccion, World)

#Agrego las conecciones de viaje
agregar_conexiones_mapa_viaje(conexiones_viaje, World)

# CONSULTA PORCENTAJE D EINFECTADOS MAYOR A 
def porcentaje_de_infectados_en_T_mayor_a(Mundo, T, infectados_mayor_a):
    infectados_mayor_a = infectados_mayor_a/100
    for i in range(T):
        pais_porcentaje = dict()
        for pais in Mundo.paises.values():
            if pais.cantidad_de_infectados_en_T():
                numero, nombre_pais, new_porcentaje = pais.cantidad_de_infectados_en_T()
                pais_porcentaje[nombre_pais] = new_porcentaje
        
        for pais in Mundo.paises.values():
            if pais.nombre in pais_porcentaje.keys():
                pais.porcentaje_infectados = pais_porcentaje[pais.nombre]
    
    a = list()
    for pais in Mundo.paises.values():
        if pais.porcentaje_infectados >= infectados_mayor_a:
            a.append(pais)
    for pays in a:
        print(pays.nombre, "--> %{}".format(round(pays.porcentaje_infectados*100),2))
    return T
    

#CONSULTA X PORCIENTO INFECTADO
def primer_pais_que_xporciento_fue_infectado(Mundo, P_inf, infect_0):
    P_inf = P_inf/100
    a = False
    pais_porcentaje = dict()
    t = 0
    while not a:
        t +=1
        for pais in Mundo.paises.values():
            if pais.cantidad_de_infectados_en_T():
                numero, nombre_pais, new_porcentaje = pais.cantidad_de_infectados_en_T()
                pais_porcentaje[nombre_pais] = new_porcentaje
        
        for pais in Mundo.paises.values():
            if pais.nombre in pais_porcentaje.keys():
                pais.porcentaje_infectados = pais_porcentaje[pais.nombre]
                
        for pais in Mundo.paises.values():
            if (pais.porcentaje_infectados*pais.habitantes-infect_0[pais.nombre])/(pais.habitantes - infect_0[pais.nombre]) >= P_inf:
                print("{}, fue el primer pais en infectarse con {}%, en T = {}".format(pais.nombre,P_inf*100, t))
                a = True
    

        
def mapa_viaje_final_con_class(mapa_viaje, Mundo):
    mapa_viaje_final = {}
    for pais in mapa_viaje:
        mini_mapa = defaultdict(list)
        if pais == Mundo.paises[pais].nombre:
            new_pais = Mundo.paises[pais]
        for conexion in mapa_viaje[pais]:
            if conexion == Mundo.paises[conexion].nombre and Mundo.paises[conexion].porcentaje_infectados < 0.2:
                conecction_class = Mundo.paises[conexion]
                mini_mapa[new_pais].append(conecction_class)
            else: 
                print(Mundo.paises[conexion])
        mapa_viaje_final.update(mini_mapa)
        
    print(mapa_viaje_final)

def mapa_viaje_final_sin_class(mapa_viaje, Mundo):
    mapa_viaje_final = {}
    for pais in mapa_viaje:
        mini_mapa = defaultdict(dict)
        if pais == Mundo.paises[pais].nombre:
            new_pais = Mundo.paises[pais].nombre
        for conexion in mapa_viaje[pais]:
            if Mundo.paises[conexion].porcentaje_infectados >= 0:
                conecction_no_class = {Mundo.paises[conexion].nombre:1}
                mini_mapa[new_pais].update(conecction_no_class)
        mapa_viaje_final.update(mini_mapa)
        
    return mapa_viaje_final
    


diccionario_conexiones = {}
for country in World.paises:
    diccionario_conexiones[country] = {}
for pais in World.paises:
    for conexion in World.paises[pais].conexiones_infeccion[pais]:
        if conexion.tasa_contagio > World.paises[pais].tolerancia:
            d = {conexion.nombre : True}
            diccionario_conexiones[pais].update(d)
        else:
            d = {conexion.nombre : False}
            diccionario_conexiones[pais].update(d)
            
diccionario_infectados_0 = {}
for pais in World.paises.values():
    diccionario_infectados_0[pais.nombre] = pais.porcentaje_infectados*pais.habitantes
    

def get_shortest_path(weighted_graph, start, end):
    """
    Calculate the shortest path for a directed weighted graph.

    Node can be virtually any hashable datatype.

    :param start: starting node
    :param end: ending node
    :param weighted_graph: {"node1": {"node2": "weight", ...}, ...}
    :return: ["START", ... nodes between ..., "END"] or None, if there is no
             path
    """

    # We always need to visit the start
    nodes_to_visit = {start}
    visited_nodes = set()
    # Distance from start to start is 0
    distance_from_start = {start: 0}
    tentative_parents = {}

    while nodes_to_visit:
        # The next node should be the one with the smallest weight
        current = min(
            [(distance_from_start[node], node) for node in nodes_to_visit])[1]

        # The end was reached
        if current == end:
            break

        nodes_to_visit.discard(current)
        visited_nodes.add(current)

        edges = weighted_graph[current]
        unvisited_neighbours = set(edges).difference(visited_nodes)
        for neighbour in unvisited_neighbours:
            neighbour_distance = distance_from_start[current] + \
                                 edges[neighbour]
            if neighbour_distance < distance_from_start.get(neighbour,
                                                            float('inf')):
                distance_from_start[neighbour] = neighbour_distance
                tentative_parents[neighbour] = current
                nodes_to_visit.add(neighbour)

    return _deconstruct_path(tentative_parents, end)


def _deconstruct_path(tentative_parents, end):
    if end not in tentative_parents:
        return None
    cursor = end
    path = []
    while cursor:
        path.append(cursor)
        cursor = tentative_parents.get(cursor)
    return list(reversed(path))


#CONSULTA A STAR
def AStar_short_route(start, end, T, World, mapa_viaje):
    World.iterar_sobre_Mundo(T)
    graph = mapa_viaje_final_sin_class(mapa_viaje, World)
    print(get_shortest_path(graph, start, end))



            
conexiones_infeccion.close()
conexiones_viaje.close()
paises.close()
```
