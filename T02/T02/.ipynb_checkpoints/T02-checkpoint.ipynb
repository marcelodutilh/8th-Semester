```python
from collections import defaultdict
conexiones_infeccion = open("conexiones_infecciÃ³n.csv","r")
conexiones_viaje = open("conexiones_viaje.csv","r")
paises = open("paises.csv","r")





class Tierra:
    
    def __init__(self):
        self.paises = {}
    
    def imprimir_mundo(self):
        print("______LISTA DE PAISES______")
        for pais in self.paises:
            print(pais)
        print("______.______")
        
class Pais:
    
    def __init__(self, nombre, habitantes, tolerancia, subterraneo, porcentaje_infectados):
        self.nombre = nombre
        self.habitantes = int(habitantes)
        self.tolerancia = float(tolerancia)
        self.subterraneo = subterraneo
        self.porcentaje_infectados = float(porcentaje_infectados)/100
        self.conexiones_infeccion = defaultdict(list)
        self.conexiones_viaje = defaultdict(list)
    
    @property
    def tasa_contagio(self):
        x = self.porcentaje_infectados*self.habitantes
        y = (1-(self.porcentaje_infectados))*self.habitantes
        z = self.habitantes
        tasa_contagio = (x-y)/z
        
        return tasa_contagio
    
    def cantidad_de_infectados_en_T(self):
        
        infectado = self.habitantes * self.porcentaje_infectados
        sanos = self.habitantes * (1-self.porcentaje_infectados)
        tasa_contagio = self.tasa_contagio
        sumatoria_contagios = 0
        Q_vecinos = 0
        for vecino in self.conexiones_infeccion[self.nombre]:
            sumatoria_contagios += vecino.tasa_contagio
            Q_vecinos +=1
        
        infectados_en_T = infectado + sanos*tasa_contagio + sanos*(1-tasa_contagio)*sumatoria_contagios/(100*Q_vecinos)
        if infectados_en_T >= 0:
            return infectados_en_T
        else:
            return 0
        
    def __repr__(self):
        #return "{}, {}, {}, {}, {}".format(self.nombre, self.habitantes, self.tolerancia, self.subterraneo, self.porcentaje_infectados)
        return self.nombre       
def crear_tierra(paises):
    Mundo = Tierra()
    arch_paises = paises
    lista_paises = list()
    for pais in arch_paises:
        new_country = pais.split(";")
        a,b,c,d,e = new_country
        if d == "TRUE":
            j =  True
        else:
            j = False
        f = e.replace("\n","")
        new_country = [a,b,c,j,f]
        lista_paises.append(new_country)
    lista_paises = lista_paises[1:]
    for pais in lista_paises:
        nombre, habitantes, tolerancia, subterraneo, porcentaje_infectados = pais
        country = Pais(nombre, habitantes, tolerancia, subterraneo, porcentaje_infectados)
        Mundo.paises[country.nombre] = country
    return Mundo

def agregar_conexiones_infecciosas(conexiones_infeccion, Mundo):
    lista_infecciones = list()
    arch_infeccion = conexiones_infeccion
    for conexion in arch_infeccion:
        if "Pais 1" in conexion:
            continue
        pais1, pais2 = conexion.split(";")
        pais_1 = pais1.replace("\n","")
        pais_2 = pais2.replace("\n","")
        Mundo.paises[pais_1].conexiones_infeccion[pais_1].append(Mundo.paises[pais_2])

def agregar_conexiones_mapa_viaje(conexiones_viaje, Mundo):
    lista_viaje = list()
    arch_viaje = conexiones_viaje
    for conexion in arch_viaje:
        if "Pais 1" in conexion:
            continue
        pais1, pais2 = conexion.split(";")
        pais_1 = pais1.replace("\n","")
        pais_2 = pais2.replace("\n","")
        Mundo.paises[pais_1].conexiones_viaje[pais_1].append(Mundo.paises[pais_2])
        

#Creo la tierra
World = crear_tierra(paises)

#Agrego las conecciones infecciosas
agregar_conexiones_infecciosas(conexiones_infeccion, World)

#Agrego las conecciones de viaje
agregar_conexiones_mapa_viaje(conexiones_viaje, World)



conexiones_infeccion.close()
conexiones_viaje.close()
paises.close()
```
